# translate function

```python
romaji_strs = list(reversed(sorted(katakana_by_char.keys() + hiragana_by_char.keys(), key=lambda x: len(x))))

def to_kana(word, verbose=False):
    for romaji_str in romaji_strs:
        hiragana, katakana = hiragana_by_char.get(romaji_str), katakana_by_char.get(romaji_str)
        if katakana:
            if verbose: print(romaji_str)
            word = word.replace(romaji_str, katakana)
            if verbose: print(word)
        else:
            if verbose: print(romaji_str)
            word = word.replace(romaji_str, hiragana)
            if verbose: print(word)
    return word

to_kana('Shitiikarejji'.lower(), True)
```

# app

```python
from flask import Flask

app = Flask(__name__)

VERBOSE = True

@app.route('/<word>')
def hello(word):
    word = str(word).lower()
    kana = to_kana(word, VERBOSE)
    return kana

app.run()
```
